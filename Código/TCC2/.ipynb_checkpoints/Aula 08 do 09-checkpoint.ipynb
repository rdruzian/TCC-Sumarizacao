```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
fs = 1000
f1 = 100
f2 = 20
f3 = 35
```

```python
t = np.arange(1000) / fs
```

```python
s1 = np.sin(2*np.pi*f1*t)
plt.plot(t,s1)
plt.show()
```

```python
s2 = np.sin(2*np.pi*f2*t)
plt.plot(t, s2)
plt.show()
```

```python
s3 = np.sin(2*np.pi*f3*t)
plt.plot(t, s2)
plt.show()
```

```python
sinal = s1 + s2 + s3
plt.plot(t, sinal)
```

```python
fc = 30
fn = fc/(fs/2)
b, a = signal.butter(3, fn, 'low')
filtrado = signal.filtfilt(b, a, sinal)
plt.plot(t, filtrado)
plt.show()
```

```python
b, a = signal.butter(10, fn, 'low')
filtrado = signal.filtfilt(b, a, sinal)
plt.plot(t, filtrado)
plt.show()
```

```python
b, a = signal.butter(16, fn, 'low')
filtrado = signal.filtfilt(b, a, sinal)
plt.plot(t, filtrado)
plt.show()
```
