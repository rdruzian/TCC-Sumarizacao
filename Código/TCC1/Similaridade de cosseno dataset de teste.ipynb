```python
import numpy as np
import pandas as pd
import nltk
from nltk import tokenize
from nltk.corpus import stopwords
from sklearn.metrics.pairwise import cosine_similarity
from gensim.models import KeyedVectors
from string import punctuation
```

```python
nltk.download('all')
```

```python
data = pd.read_json('tcc1.json')
```

```python
noticia = []
titulo = []
for t in range(len(data.título)):
    titulo.append(data.título[t].lower())
for n in range(len(data.texto)):
    noticia.append(data.texto[n].lower())
```

```python
noticia_frase = []
titulo_frase = []
for t in range(len(titulo)):
    titulo_frase.append(titulo[t].split("."))
for n in range(len(noticia)):
    noticia_frase.append(noticia[n].split("."))
```

```python
token_noticias = []
token_titulos = []
for t in range(len(titulo)):
    token_titulos.append(nltk.word_tokenize(titulo[t]))
for n in range(len(noticia)):
    token_noticias.append(nltk.word_tokenize(noticia[n]))
```

```python
pontos = []
for ponto in punctuation:
    pontos.append(ponto)
```

```python
stopwords = nltk.corpus.stopwords.words("portuguese")
```

```python
stopwords = stopwords + pontos
```

```python
model = KeyedVectors.load_word2vec_format('glove_s100.txt', encoding='utf-8')
```

Remoção de stopwords

```python
titulo_limpo = []
for i in range(len(token_titulos)):
    frase = []
    token = token_titulos[i]
    for j in range(len(token)):
        if token[j] not in stopwords and token[j] not in pontos:
            frase.append(token[j])
    titulo_limpo.append(' '.join(frase))
```

```python
noticia_limpa = []
for i in range(len(token_noticias)):
    frase = []
    token = token_noticias[i]
    for j in range(len(token)):
        if token[j] not in stopwords and token[j] not in pontos:
            frase.append(token[j])
    noticia_limpa.append(' '.join(frase))
```

Craição do dicionário de embbedings

```python
titulo_embbeding = {}
for t in token_titulos:
    for j in range(len(t)):
        palavra = t[j]
        if palavra in model:
            titulo_embbeding[palavra] = model[palavra]
```

```python
noticia_embbeding = {}
for t in token_noticias:
    for j in range(len(t)):
        palavra = t[j]
        if palavra in model:
            noticia_embbeding[palavra] = model[palavra]
```

Criado variáveis com a média das embbedings de cada título

```python
media_titulos = []
for t in range(len(token_titulos)):
    m = 0
    for j in token_titulos[t]:
        if j in titulo_embbeding:
            m += titulo_embbeding[j]
    media_titulos.append(m/len(token_titulos[t]))
```

Criado vetor com o valor escalar para cada título

```python
media_titulo_escalar = []
for m in range(len(media_titulos)):
    media = 0
    for i in range(len(media_titulos[m])):
        media += media_titulos[m][i]
    media_titulo_escalar.append(media/100)
```

Criado variáveis com a média das embbedings de cada notícia

```python
media_noticias = []
for t in range(len(token_noticias)):
    m = 0
    for j in token_noticias[t]:
        if j in noticia_embbeding:
            m += noticia_embbeding[j]
    media_noticias.append(m/len(token_noticias[t]))
```

Criado vetor com o valor escalar para cada notícia

```python
media_noticia_escalar = []
for m in range(len(media_noticias)):
    media = 0
    for i in range(len(media_noticias[m])):
        media += media_noticias[m][i]
    media_noticia_escalar.append(media/100)
```

```python
similaridade_matriz = np.zeros((len(media_titulos),len(media_noticias)), dtype=np.float64)
for i in range(len(media_titulos)):
    for j in range(len(media_noticias)):
        #if i != j:
        similaridade_matriz[i][j] = cosine_similarity(media_titulos[i].reshape(1,100), media_noticias[j].reshape(1,100))[0,0]
```

```python
for i in range(len(media_titulos)):
    maior = 0
    menor = 0
    for j in range(len(media_noticias[i])):    
        if similaridade_matriz[i][j] > maior:
            maior = j
        if similaridade_matriz[i][j] < menor:
            menor = j
    with open("resultado.txt", "+a", encoding='utf-8') as file:
        file.write("Título: " + titulo[i] + "\n" + "Noticia mais semelhante: " + noticia_frase[i][maior] + "\n" + 
               "Noticia menos semelhante: " + noticia_frase[i][menor] + "\n")
```
