```python
import numpy as np
import pandas as pd
import nltk
nltk.download('punkt')
import re
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.model_selection import train_test_split
import networkx as nx
from gensim.models import KeyedVectors
```

```python
data = pd.read_json("tcc1.json")
```

```python
#Lista de frases
frase = []
#Para cada texto, será feito um split no '.' e preenche a lista frase
for f in data['texto']:
  frase.append(sent_tokenize(f))
```

```python
frase = [y for x in frase for y in x]
len(frase)
```

```python
frase = frase[:1000]
```

Vetorização de palavras

```python
vetor_embbeding = {}
model_glove = open('glove.6B.100d.txt', encoding='utf-8')
for l in model_glove:
    valores = l.split()
    word = valores[0]
    coefs = np.asarray(valores[1:], dtype='float32')
    vetor_embbeding[word] = coefs

model_glove.close()
```

```python
len(vetor_embbeding)
```

Em frase_limpa irá ter o conteúdo de frase mas sem pontuações ou caracteres especiais.<br>
Após isso é transformado todo o texto para letras minusculas

```python
frase_limpa = pd.Series(frase).str.replace("[^a-zA-Z]", " ")
frase_limpa = [f.lower() for f in frase_limpa]
```

```python
nltk.download('stopwords')
```

```python
stop_words = stopwords.words('portuguese')
```

```python
def remove_stopwords(sen):
    nova_frase = " ".join([i for i in sen if i not in stop_words])
    return nova_frase
```

```python
#Das frases sem caracteres especiais e em minusculo, são removidos os stops-words
frase_limpa = [remove_stopwords(r.split()) for r in frase_limpa]
```

Agora iremos criar os vetores das frases

```python
frase_embbeding = {}
model_glove = open('glove.6B.100d.txt', encoding='utf-8')
for l in model_glove:
    valores = l.split()
    word = valores[0]
    coefs = np.asarray(valores[1:], dtype='float32')
    frase_embbeding[word] = coefs

model_glove.close()
```

```python
vetor_frase = []
for i in frase_limpa:
  if len(i) != 0:
    v = sum([frase_embbeding.get(w, np.zeros((100,))) for w in i.split()])/(len(i.split())+0.001)
  else:
    v = np.zeros((100,))
  vetor_frase.append(v)
```

```python
similaridade_matriz = np.zeros([len(frase), len(frase)])
```

```python
len(frase)
```

```python
for i in range(len(frase)):
    for j in range(len(frase)):
        if i != j:
            similaridade_matriz[i][j] = cosine_similarity(vetor_frase[i].reshape(1,100), vetor_frase[j].reshape(1,100))[0,0]
    print(str(i) + " de " + str(len(frase)))
```

```python
grafo_nx = nx.from_numpy_array(similaridade_matriz)
scores = nx.pagerank(grafo_nx)
```

```python
frases_ranqueadas = sorted(((scores[i],s) for i,s in enumerate(frase)), reverse=True)
```
